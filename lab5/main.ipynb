1. Opisz krótko EEG. W jakim celu stosuje się to badanie?

EEG polega na nieinwazyjnym uzyskaniu zapisu czynności elektrycznej kory mózgu, która po odpowiedniej interpretacji może stanowić 
bardzo pomocne narzędzie w rozpoznaniu wielu zaburzeń neurologicznych. Badanie EEG swoimi korzeniami sięga końca XIX wieku, od tego 
czasu było rozwijane i doskonalone przez wielu wybitnych badaczy i niewątpliwie przyczyniło się do lepszego zrozumienia funkcjonowania
ludzkiego mózgu.

2. Wczytaj i narysuj plik reprezentujący EEG o nazwie zdrowy.dat, który posiada 12 odprowadzeń:
'Fp1', 'Fp2', 'F3', 'F4', 'C3', 'C4', 'T3', 'T4', 'P3', 'P4', 'O1', 'O2'.

```python
channels = ['Fp1', 'Fp2', 'F3', 'F4', 'C3', 'C4', 'T3', 'T4', 'P3', 'P4', 'O1', 'O2']
```

```python
def draw_plot(data):
    fig, axs = plt.subplots(3, 4, figsize=(16, 12))
    fig.suptitle('Time Series Plots for EEG Channels', fontsize=16)
    
    for i, channel in enumerate(channels):
        row, col = i // 4, i % 4 
        axs[row, col].plot(data[channel], color='blue')
        axs[row, col].set_title(channel)
        axs[row, col].set_xlabel('Time')
        axs[row, col].set_ylabel('Amplitude')
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
```

```python
data_zdrowy = np.genfromtxt('sygnalyEEG/zdrowy.dat',
                     skip_header=1,
                     skip_footer=1,
                     names=True,
                     dtype=int,
                     delimiter='  ')

data_zdrowy = data_zdrowy[:2048]
data_zdrowy.dtype.names = channels

data_zdrowy[:5]
```

```python
draw_plot(data_zdrowy)
```

3. Wczytaj i narysuj plik reprezentujący EEG o nazwie bol.dat, który posiada 12 odprowadzeń:
'Fp1', 'Fp2', 'F3', 'F4', 'C3', 'C4', 'T3', 'T4', 'P3', 'P4', 'O1', 'O2'.

```python
data_chory = np.genfromtxt('sygnalyEEG/bol.dat',
                     skip_header=1,
                     skip_footer=1,
                     names=True,
                     dtype=int,
                     delimiter='  ')

data_chory = data_chory[:2048]
data_chory.dtype.names = channels

data_chory[:5]
```

```python
draw_plot(data_chory)
```
